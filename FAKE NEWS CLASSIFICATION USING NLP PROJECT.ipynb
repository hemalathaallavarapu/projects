```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
index = []
corpus = []
```

```python
for i in range(0, len(messages)):
  try:
    review = re.sub('[^a-zA-Z]', " ",messages['title'][i])
    review = review.lower()
    review = review.split()

    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]

    review = " ".join(review)

    corpus.append(review)

    index.append(i)
  except:
    pass
```

```python
corpus
```

```python
y = data['label'].loc[index]
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf_v = TfidfVectorizer(max_features=5000, ngram_range=(1,3))
X = tfidf_v.fit_transform(corpus).toarray()
```

```python
X
```

```python
data2 = pd.DataFrame()
```

```python
data2['corpus'] = corpus
```

```python
vectorized_data = pd.DataFrame(X, columns=tfidf_v.get_feature_names_out())
```

```python
vectorized_data
```

```python
vectorized_data['zone'][vectorized_data['zone']!=0]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(vectorized_data, y, test_size=0.2, random_state=42)
```

```python
from sklearn.naive_bayes import BernoulliNB, MultinomialNB, GaussianNB

from sklearn import metrics
```

```python
model1 = BernoulliNB()
model1.fit(x_train, y_train)
```

```python
ans1 = model1.predict(x_test)
```

```python
metrics.accuracy_score(y_test, ans1)*100
```

```python
data.label.value_counts(
)
```

```python
print(metrics.classification_report(y_test, ans1))
```

```python
import seaborn as sns
```

```python
sns.heatmap(metrics.confusion_matrix(y_test,ans1), annot=True, fmt='.1f')
```
